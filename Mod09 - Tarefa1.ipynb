### 2. Peso médio dos bebes por sexo que nasceram no dia do seu aniversário por faixas de escolaridade mae
Ex: Você, aluna(o), nasceu no dia 10/01, então você precisa filtrar o conjunto de dados nessa data e calcular o peso médio dos bebês de cada sexo por faixa de escolaridade da mãe.

```python
# Selecionando só as colunas que preciso

sinasc_2 = sinasc_raw[['PESO', 'SEXO', 'DTNASC', 'ESCMAE']]
sinasc_2.head()
```

```python
# Criei um nova DF só com 'DTNASC' == '2019-10-16'

niver = sinasc_2[sinasc_2['DTNASC'] == '2019-10-16']
niver
```

```python
# Peso médio dos bebes por sexo que nasceram no meu niver por faixas de escolaridade mae

niver.groupby(['ESCMAE', 'SEXO'])['PESO'].mean().unstack()
```

```python
# Mesmo comando, porém sem unstack. 
# Observe que não exibe dados para Masculino na ESCMAE = 1 a 3 anos, confirmando o "NaN" acima.

niver.groupby(['ESCMAE', 'SEXO'])['PESO'].mean()
```

```python
# Como exemplo, executando o mesmo comando, porém consultando tada a base

sinasc_2.groupby(['ESCMAE', 'SEXO'])['PESO'].mean()
```

### 4. Qual o municipio que nasceu mais bebe no mês de março?
    a) qual a quantidade de filhos vivos media, maxima, minima nesse municipio?
    b) qual a idade media, maxima, minima dos pais nesse municipio?



```python
#Criando DF com os dados do mês de março

selecao = (sinasc_raw['DTNASC'] >= '2019-03-01') & (sinasc_raw['DTNASC'] <= '2019-03-31')
marco = sinasc_raw[selecao]
```

```python
marco
```

```python
# Consultando qual município teve mais nascimentos - Essa consulta exibe só o valor

marco.munResNome.value_counts().max()
```

```python
# Consultando qual município teve mais nascimentos

marco.munResNome.value_counts()
```

### Qual o municipio que nasceu mais bebe no mês de março?
    - Porto Velho

```python
#Criando DF somente com o município com mais nascimentos no mês de março

selecao = (marco['munResNome'] == 'Porto Velho') & (marco['DTNASC'])
marco_2 = marco[selecao]
marco_2
```

```python
# Qual a quantidade de filhos vivos media, maxima, minima nesse municipio?
```

```python
marco_2.groupby('munResNome')['QTDFILVIVO'].count()
```

```python
tarefa_4 = marco_2[['QTDFILVIVO', 'DTNASC']].copy()
tarefa_4
```

```python
resultado = tarefa_4.groupby('DTNASC')['QTDFILVIVO'].count()
resultado
```

```python
# Média da quantidade de filhos vivos por dia do mês de março ordenado pela data de nascimento

tarefa_4.groupby('DTNASC', sort=True)['QTDFILVIVO'].mean().to_frame()
```

```python
# Média, Mínimo e Máximo de fihos vivos por dia do mês de março

agrupamento = tarefa_4.groupby(['DTNASC'])
agrupamento.agg([('Média','mean'), ('Máximo','max'), ('Mínimo','min')])
```
