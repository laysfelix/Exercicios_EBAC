```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esse arquivo.

```python
#df_sinasc = pd.read_csv('https://xxx/RO_2019.csv')
#print(df_sinasc.info())
#df_sinasc

nascidos_RO2019 = pd.read_csv('RO_2019.csv')
nascidos_RO2019
```

#### 2) Categorize a variável APGAR5 conforme as seguintes quebras:

- 0 a 3: 'asfixia grave'
- 4 a 5: 'asfixia moderada'
- 6 a 7: 'asfixia leve'
- 8 a 10: 'normal'

```python
# Criando as categorias

nascidos_RO2019.loc[(nascidos_RO2019['APGAR5'] >= 0) & (nascidos_RO2019['APGAR5'] <= 3),
           'APGAR_CAT'] = 'ASFIXIA GRAVE'
nascidos_RO2019.loc[(nascidos_RO2019['APGAR5'] >= 4) & (nascidos_RO2019['APGAR5'] <= 5),
           'APGAR_CAT'] = 'ASFIXIA MODERADA'
nascidos_RO2019.loc[(nascidos_RO2019['APGAR5'] >= 6) & (nascidos_RO2019['APGAR5'] <= 7),
           'APGAR_CAT'] = 'ASFIXIA LEVE'
nascidos_RO2019.loc[(nascidos_RO2019['APGAR5'] >= 8) & (nascidos_RO2019['APGAR5'] <= 10),
           'APGAR_CAT'] = 'NORMAL'
```

```python
nascidos_RO2019[['APGAR5', 'APGAR_CAT']]
```

```python
# Criando o gráfico de barras exibindo as categorias - distribuição do minuto por categoria

sns.countplot(x='APGAR_CAT', orient = 'v'
              , hue = 'APGAR5'
              , data=nascidos_RO2019[nascidos_RO2019.APGAR5 < 10]
             )
```

```python
# Criando o gráfico do tipo barras exibindo as categorias - distribuição de categoria por minuto

sns.countplot(x='APGAR5', orient = 'V'
              , hue = 'APGAR_CAT'
              , data=nascidos_RO2019[nascidos_RO2019.APGAR5 < 10]
             )
```

#### 3) Monte um gráfico comparando a distribuição de APGAR5 por sexo para cada grupo de tempo de gestação (GESTACAO).

```python
# usando o seaborn - CERTO MAS ARRUMAR A SEMANAS
ax = sns.boxplot(data = nascidos_RO2019, y="APGAR5", x="GESTACAO", orient="v", hue='SEXO', width=0.9)
ax.set_xticklabels(labels=['37 a 41 semanas', '42 semanas e mais', '32 a 36 semanas', '22 a 27 semanas', '28 a 31 semanas', 'Menos de 22 semanas'], rotation=30)
#ax.set_xticklabels(labels=[1, 2, 3, 4, 5, '42 semanas e mais'], rotation=30)
```

####  4) Monte uma tabela cruzada com APGAR5 nas colunas e GESTACAO nas linhas

```python
nascidos = pd.crosstab(nascidos_RO2019['GESTACAO'], nascidos_RO2019['APGAR5'])
nascidos
```

####  5) Da tabela cruzada do intem anterior, calcule os percentuais de linha. Ou seja, a distribuição da variável APGAR5 para cada classe de GESTACAO dado.

```python
nascidos.sum()
```

```python
nascidos.sum(axis = 1)
```

```python
# distribuição da variável APGAR5 para cada classe de GESTACAO

pct_gestacao = nascidos.div(nascidos.sum(axis = 1), axis = 0)
pct_gestacao
```

####  6) Os médicos costumam classificar parto prematuro quando a gestação tem menos que 37 semanas. Faça um gráfico de barras da tabela cruzada acima e conclua se você acha que este conjunto de dados corrobora ou não com a afirmação dos médicos.

```python
# Gráfico de barra da tabela cruzada

nascidos = pd.crosstab(nascidos_RO2019['GESTACAO'], nascidos_RO2019['APGAR5'])
print(nascidos)
print(nascidos.sum(1))

pct_gestacao = nascidos.div(nascidos.sum(axis = 1), axis = 0)
print(pct_gestacao)

pct_gestacao.plot.bar()
```

Resposta da questão 6: Não confirma a afirmação dos médicos, pois o gráfico acima demonstra que os nascimentos prematuros tem a pontuação menor (axfixia moderada a grave) no APGAR5 que são os primeiros 5 minutos de vida, comparando com os nacimentos após a 37º semana de gestação que tem a pontuação maior (axfixia leve a normal).
