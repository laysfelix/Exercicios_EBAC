#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esses arquivos.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#%matplotlib inline 
#%matplotlib notebook
```

```python
#df_sinasc = pd.read_csv('https://xxx/RO_2019.csv')
#print(df_sinasc.info())
#df_sinasc

nascidos_RO2019 = pd.read_csv('RO_2019.csv')
nascidos_RO2019
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
plt.rc('figure', figsize=(5, 5))
```

```python
nascidos_RO2019['APGAR1'].value_counts().sort_index().plot.bar()
plt.title(label="'APGAR1'")
plt.xlabel(xlabel='APGAR no 1º minuto')
plt.tick_params(axis='x', rotation=0)
plt.show()
```

```python
nascidos_RO2019['APGAR5'].value_counts().sort_index().plot.bar()
plt.title(label="'APGAR5'")
plt.xlabel(xlabel='APGAR no 5º minuto')
plt.tick_params(axis='x', rotation=0)
plt.show()
```

```python
nascidos_RO2019['SEXO'].value_counts().sort_index().plot.bar()
plt.title(label="'SEXO'")
plt.tick_params(axis='x', rotation=0)
plt.show()
```

```python
nascidos_RO2019['GESTACAO'].value_counts().reindex(['Menos de 22 semanas', '22 a 27 semanas', '28 a 31 semanas', 
                                                    '32 a 36 semanas', '37 a 41 semanas', '42 semanas e mais']).plot.bar()
plt.title(label="'GESTACAO'")
plt.xlabel(xlabel='Semanas da gestação')
plt.tick_params(axis='x', rotation=45)
plt.show()
```

```python
nascidos_RO2019['ESTCIVMAE'].value_counts().sort_index().plot.bar()
plt.title(label="'ESTCIVMAE'")
plt.xlabel(xlabel='\n Estado civil: \n\n1: Solteira \n2: Casada \n3: Viuva \n4: Separado judicialmente/Divorciado '
           '\n5: União consensual \n9: Ignorado')
plt.tick_params(axis='x', rotation=45)
plt.show()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
print('Apgar no 5º minuto: \n0 a 10 \n')
print('Número de consultas de pré-natal: \n1: Nenhuma \n2: de 1 a 3 \n3: de 4 a 6 \n4: 7 e mais \n9: Ignorado')

plt.rc('figure', figsize=(5, 5))
ax = sns.barplot(data=nascidos_RO2019, 
                 x='APGAR5', 
                 y='CONSULTAS')
ax.set_yticks(ticks=[1, 2, 3, 4])
plt.xlabel(xlabel="\n APGAR5: Apgar no 5º minuto")

plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
plt.rc('figure', figsize=(16, 16)) # tamanho da imagem
fig, axes = plt.subplots(2, 2) # 2 colunas e 2 linhas
fig.subplots_adjust(hspace=0.4)

ax1 = sns.countplot(ax = axes[0, 0], x='SEXO', data=nascidos_RO2019)

ax2 = sns.countplot(ax = axes[0, 1], x='GESTACAO', data=nascidos_RO2019, order=['Menos de 22 semanas', 
                           '22 a 27 semanas', 
                           '28 a 31 semanas', 
                           '32 a 36 semanas', 
                           '37 a 41 semanas', 
                           '42 semanas e mais'])
ax2.set_xticklabels(labels=ax2.get_xticklabels(), 
                    rotation=45)
ax2.set_xlabel(xlabel='Semanas de gestação')

ax3 = sns.countplot(ax = axes[1, 0], x='ESTCIVMAE', data=nascidos_RO2019)
ax3.set_xticklabels(labels=['1: Solteira', 
                            '2: Casada', 
                            '3: Viúva', 
                            '4: Separado judicialmente\nDivorciado', 
                            '5: União consensual',
                            '9: Ignorado'], 
                    rotation=45)
ax3.set_xlabel(xlabel='Estado civil')

ax4 = sns.countplot(ax = axes[1, 1], x='CONSULTAS', data=nascidos_RO2019)
ax4.set_xticklabels(labels=['1: Nenhuma', 
                            '2: de 1 a 3', 
                            '3: de 4 a 6', 
                            '4: 7 e mais', 
                            '9: Ignorado'], 
                    rotation=45)
ax4.set_xlabel(xlabel='Número de consultas de pré-natal')

plt.show()
```
