````python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
````

### Carregando o dataframe <a name="read"></a>
<div style="text-align: right"
     
[Voltar ao índice](#Contents)

````python
previsao_de_renda = pd.read_csv('previsao_de_renda.csv')
previsao_de_renda.head()
````

````python
var_qual = sns.pointplot(x="posse_de_veiculo", y="renda",
                   data=previsao_de_renda, dodge=True, errorbar=('ci', 90))
````

````python
var_qual = sns.pointplot(x="posse_de_imovel", y="renda",
                   data=previsao_de_renda, dodge=True, errorbar=('ci', 90))
````

````python
As variáveis posse de imóvel e posse de veículo são boas preditoras 
de renda, pois quem tem a renda maior possuem veículo e/ou imóvel.
````

````python
ax = sns.pointplot(x="posse_de_imovel", y="renda", hue = 'sexo',
                   data=previsao_de_renda, dodge=True, errorbar=('ci', 95))

plt.legend(loc = 'lower center', bbox_to_anchor=(0.5, -.30),ncol = 3)
````

````python
ax = sns.pointplot(x="posse_de_veiculo", y="renda", hue = 'sexo',
                   data=previsao_de_renda, dodge=True, errorbar=('ci', 95))

plt.legend(loc = 'lower center', bbox_to_anchor=(0.5, -.30),ncol = 3)
````

As variáveis posse de imóvel e posse de veículo não são boas preditoras 
de renda quando é avaliado a variável Sexo junto. 
O sexo masculino demonstra ter a renda mais alta, apesar da minoria possuir veículo e/ou imóvel.
