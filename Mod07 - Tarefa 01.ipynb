#### 4) Qual variável é mais poderosa?

Considere as variáveis ```possui_email``` e ```posse_de_veiculo```. Faça uma tabela cruzada entre elas e responda qual delas te parece mais poderosa para prever a probabilidade de ```mau = 1```?

````python
dados_cliente.mau.unique()
````

````python
dados_cliente.possui_email.unique()
````

````python
dados_cliente.posse_de_veiculo_Y.unique()
````

````python
#Tabela cruzada das variáveis mau pagador e possui veículo

mau_veiculo = pd.crosstab(dados_cliente['mau'],dados_cliente['posse_de_veiculo_Y'])
mau_veiculo = mau_veiculo.div(mau_veiculo.sum(axis=1),axis=0)
mau_veiculo

#cliente_avaliacao = pd.crosstab(dados_cliente['mau'], dados_cliente['posse_de_veiculo_Y'])
#cliente_avaliacao
````

````python
#Tabela cruzada das variáveis mau pagador e possui e-mail

mau_email = pd.crosstab(dados_cliente['mau'],dados_cliente['possui_email'])
mau_email = mau_email.div(mau_email.sum(axis=1),axis=0)
mau_email

#mau_email = pd.crosstab(dados_cliente['mau'], dados_cliente['possui_email'])
#mau_email
````

A variável <b> possui veículo </b> é ideal para prever a probabilidade de mau pagador, pois ter um veículo gera custos e despesas não organizadas torna o cliente inadimplente. <br>
Grande parte dos maus pagadores não tem <b> e-mail </b> cadastrado. O que dificultaria num contato ao cliente por e-mail.<br>
Analisando a variável <b> possui veículo</b>, verificamos que 63% dos inadimplentes não possui veículo.
